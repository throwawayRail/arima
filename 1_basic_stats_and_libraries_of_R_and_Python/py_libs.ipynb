```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_style("darkgrid")
data = np.array(["h", "e", "l", "l", "o"])
ser = pd.Series(data)
print(ser)
```

```python
 employee = [(101, 'Alex', 28, 20000),
(102, 'Megan', 32, 25000),
(103, 'Josh', 25, 40000),
(104, 'Jane', 32, 35000),
(105, 'Alice', 32, 30000),
(106, 'Sally', 32, 20000),
(107, 'Rob', 40, 24000)]
df = pd.DataFrame(employee, columns = ["Id", "Name", "Age", "Salary"])
df
```

```python
from scipy import stats
print("Mean Salary is: {}".format(np.mean(df['Salary'])))
print("Median Salary is: {}".format(np.median(df['Salary'])))
print("Mode Salary is: {}".format(stats.mode(df['Salary']).mode))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(df['Name'], df['Salary'], "b.-")
plt.xlabel("Employee Name")
plt.ylabel("Salary")
plt.title("Employee Name vs Salary")
```

```python
print("Salary variance: {}".format(np.var(df['Salary'])))
print("Salary Standard Deviation: {}".format(np.std(df['Salary'])))
```

```python
# Years (x-axis)
years = [2010, 2012, 2014, 2016, 2018, 2020]
# Electricity consumption data for India & Pakistan (y-axis)
indiaConsumption = [1200, 1400, 1600, 1800, 2000, 2200]
pakistanConsumption = [800, 1000, 1200, 1400, 1600, 1800]
plt.plot(years, indiaConsumption, 'ro--', label = 'India')
plt.plot(years, pakistanConsumption, '^-', color = 'orange', linestyle ='dashed', label = 'Pakistan')
plt.xlabel("Years")
plt.ylabel("Electricity Consumption")
plt.title("Electricity Consumption of India and Pakistan over the Years")
plt.legend()
```

```python
 dataset1_x = [1, 2, 3, 4, 5]
dataset1_y = [10, 12, 8, 15, 11]
dataset2_x = [2, 3, 4, 5, 6]
dataset2_y = [8, 10, 14, 12, 9]
plt.scatter(dataset1_x, dataset1_y, label = 'Dataset 1', marker = 'o', color ='green')
plt.scatter(dataset2_x, dataset2_y, label = 'Dataset 2', marker = '*', color ='purple')
plt.legend()
plt.title("Scatter Plot of 2 Datasets")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```
