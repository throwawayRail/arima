```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
x = [10, 20, 30, 40, 50]
y = [20, 30, 40, 50, 60]
```

```python
plt.plot(x, y)
```

```python
plt.title("Line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x, y)
plt.show()
```

```python
data = pd.read_csv("/kaggle/input/tipscsv/tips.csv")
x = data["day"]
y = data['total_bill']
plt.bar(x, y)
plt.title("Bar Chart")
plt.ylabel("Total Bill")
plt.xlabel("Day")
plt.show()
```

```python
x = data["total_bill"]
plt.hist(x)
plt.title("histogram")
plt.ylabel("Frequency")
plt.xlabel("Total Bill")
plt.show()
```

```python
x = data["day"]
y = data["total_bill"]
plt.scatter(x, y)
plt.title("Scatter plot")
plt.ylabel("Total bill")
plt.xlabel('Day')
plt.show()
```

```python
cars = ["AUDI", "BMW", "FORD", "TESLA", "JAGUAR"]
data = [23, 10, 35, 15, 12]
plt.pie(data, labels=cars)
plt.title("Pie Chart")
```

```python
np.random.seed(10)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, vert=True, patch_artist = True, boxprops=dict(facecolor="skyblue"),
           medianprops=dict(color="red"))
plt.xlabel("Data set")
plt.ylabel("Values")
plt.title("Box Plot")
plt.show()
```
