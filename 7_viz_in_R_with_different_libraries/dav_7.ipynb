```r
cone <- function(x, y) {
sqrt(x ^ 2 + y ^ 2)
}
x <- y <- seq(-1, 1, length=30)
z <- outer(x, y, cone)
persp(x, y, z,
main="perspective plot of cone",
zlab="height",
theta=30, phi=15,
col="orange", shade=0.4)
```
